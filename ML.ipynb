Mean Normalization 

```python
import numpy as np
features = np.random.randint(0,10,size =(2,10))
```

```python
features
```

```python
def mean_norm(X):
  return np.subtract(X,np.mean(X,axis=1).reshape(-1,1))/(np.max(X,axis=1)-np.min(X,axis=1)).reshape(-1,1)
```

```python
mean_norm(features)
```

```python
X = features
```

```python
np.mean(X,axis=1),(np.max(X,axis=1)-np.min(X,axis=1))
```

```python
np.subtract(X,np.mean(X,axis=1).reshape(-1,1))
```

```python
np.mean(X,axis=1).reshape(-1,1).shape
```

```python
# create x
x = np.asarray([1,2,3,4], dtype=np.float64)

np.mean(x) # calculates the mean of the array x
x-np.mean(x) # this is euivalent to subtracting the mean of x from each value in x
x-=np.mean(x) # the -= means can be read as x = x- np.mean(x)

np.std(x) # this calcualtes the standard deviation of the array
x/=np.std(x) # the /= means can be read as x = x/np.std(x)
```
